```python
import matplotlib.pyplot as plt
#Import the library:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.collections

plt.style.use('default')
```

```python
df = pd.read_csv('penguins.csv')
```

```python
df.head()
```

```python
plt.plot(df['species'], df['bill_length_mm'], color='Green', linestyle='dashdot', marker="D", markersize=10, label="bill_depth_mm")
plt.plot(df['species'], df['flipper_length_mm'], color='Red', linewidth=6, label='flipper_length_mm')
plt.title('species vs length')
plt.xlabel('Species')
plt.ylabel('Length')
plt.legend()
plt.show()
```

```python
plt.plot(df['island'], df['bill_length_mm'])
```

```python
df.dropna(inplace=True)
```

```python
plt.plot(df['sex'], df['bill_length_mm'])
```

```python
plt.scatter(df['flipper_length_mm'], df['bill_length_mm'], color='blue', marker="x", s=df['body_mass_g']*20)
```

```python
plt.figure(figsize=(8,4))

plt.scatter(df['bill_length_mm'],df['bill_depth_mm'],c=df['year'],cmap='magma',alpha=1.0)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.axhline(15,color='red')
plt.axhline(12,color='green')
plt.axvline(45,color='red')
plt.colorbar()
```
