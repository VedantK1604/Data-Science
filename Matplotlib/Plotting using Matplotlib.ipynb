# 2D Line Plot

2D line plot is used for bivariate analysis.
1. categorical-numerical and numerical-categorical.
2. use case -> Time series data.

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

#Categorical data on 'x' and numerical data on 'y'.
plt.plot(year, price)
#plt.plot(x,y)
```

```python
#From a pandas dataframe:
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#plotting multiple plots:
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#Label Title:
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit vs Kohli Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='pink')
plt.plot(batsman['index'],batsman['RG Sharma'],color='purple')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#line style:
plt.plot(batsman['index'],batsman['V Kohli'],color='pink',linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],color='purple',linestyle='dotted')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#Line width and marker:
plt.plot(batsman['index'],batsman['V Kohli'],color='pink',linestyle='solid',linewidth=3,marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='purple',linestyle='dotted',linewidth=3,marker='o',markersize=10)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#legend location:
plt.plot(batsman['index'],batsman['V Kohli'],color='pink',linestyle='solid',linewidth=3,marker='D',label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'],color='purple',linestyle='dotted',linewidth=3,marker='o',markersize=10,label="Rohit")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
#legend contains parameter with default value 'best' means, it will find the best location on graph to fit the legend.
```

```python
#limiting axes:
price = [48000,54000,57000,49000,47000,4500000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year, price)
```

```python
price = [48000,54000,57000,49000,47000,4500000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year, price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
#It works like zoomin to the graph.
```

```python
#Grid:
plt.plot(batsman['index'],batsman['V Kohli'],color='pink',linestyle='solid',linewidth=3,marker='D',label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'],color='purple',linestyle='dotted',linewidth=3,marker='o',markersize=10,label="Rohit")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
plt.grid()
```

```python
#show:
plt.plot(batsman['index'],batsman['V Kohli'],color='pink',linestyle='solid',linewidth=3,marker='D',label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'],color='purple',linestyle='dotted',linewidth=3,marker='o',markersize=10,label="Rohit")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
plt.grid()
#In software such as 'pycharm' we must have to use show() function. Otherwise, it will does not show the graph.
plt.show()
```

# Scatter Plot

1. Use for bivariate analysis
2. mostly works on numerical vs numerical
3. To find correlation between 2 quantities.

```python
#plt.scatter simple function:
x = np.linspace(-10,10,50)
x
```

```python
y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

```python
#On pandas dataframe:
df = pd.read_csv('batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='blue',marker='*')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

# Bar Chart

1. Bivariate Analysis
2. Numerical vs Categorical
3. use case - Aggregate analysis of group

```python
#Simple bar chart:
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='purple')
```

```python
#bar chart using data:
#colors and labels:
```

```python
#Horizontal bar chart:
plt.barh(colors,children,color='black')
```

```python
#Multiple bar charts:
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
```

```python
#Solution: Xtricks:
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Staked bar chart:
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

# Pie Chart

1. Univariate/Bivariate Analysis
2. Categorical vs numerical
3. Use case - To find contibution on a standard scale

```python
#Simple data:
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
