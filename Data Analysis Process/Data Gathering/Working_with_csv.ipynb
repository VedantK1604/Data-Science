# 3. Opening a csv file from an url / Server

When we have to load data from the online server. Code snippet is available, we have to just use it.

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv" #Url of the server to be passed
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data).head()
```

# 4. Sep Parameter

```python
#Defaultly, csv file is seperated by commas, but tsv files are seperated bt tabs, so it will never finds the commas as does not 
# split the columns. Hence, we have to use the 'sep' parameter and have to split with respect to tab '\t'.
pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sno','name','release_year','rating','votes','genres']).head()
#names: when we don't have column name in our data, pandas takes the first row by default and make it column name, so overcome 
#  this issue, we can provide column name using 'names'.
```

# 12. Handling Dates

```python
pd.read_csv('IPL Matches 2008-2020.csv',parse_dates=['date']).info()
```

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

# 14. Converters

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename}).head()
```
