Groupby objects is bacially the groups.


```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('imdb-top-1000.csv')
```

```python
movies.head()
```

```python
#we form the groups on the basis of data is actually a group.
#numerical & categorical.
#groupby works on categorical data.
```

```python
genres = movies.groupby('Genre')
genres
```

```python
#Applying builtin aggregation functions on groupby objects.
genres.sum()
# genres.mean()
#std()
#min()
#max()
#Firstly, it made the group on the basis of genre and then the sum() function runs on every numerical column.
#Eg. every movie in action, has runtime total 22196 hours.
```

```python
#Find the top 3 genres by total earning.
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
#find the genre with highest avg imdb_rating.
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
#Find director with most popularity.
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
#find the highest rated movies of each genre.
#movies.groupby('Genre')['IMDB_Rating'].max().sort_values(ascending=False)
```

```python
#find number of movies done by each actor.
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
#Groupby attributes and methods.
```

```python
#Find the total number of groups -> len.
len(movies.groupby('Genre'))
#or movies['Genre'].ununique()
```

```python
#Find items in each group -> size
movies.groupby('Genre').size()
#or (movies.groupby('Genre').value_counts()
```

```python
#first()/last()-> nth item
genres = movies.groupby('Genre')
#genres.first()-> for 1st item in every group
#genres.last()-> last item in every group
genres.nth(6) # 7th item in every group
```

```python
#get_group() -> vs filtering
#can see every item of a particular group. eg. action.
genres.get_group('Horror')
#or by filtering
# movies[movies['Genre']=='Horror']
```

```python
genres.groups #give index position of every group items.
```

```python
genres.describe()
```

```python
genres.sample() #it will take a random item from every group
#for getting more than one random value every group
#genres.sample(2,replace=True)
```

```python
#nunique -> unique values in every group.
genres.nunique()
```

```python
#agg function:
#passing dict.
genres.agg(
{
    'Runtime':'mean',
    'IMDB_Rating': 'mean',
    'No_of_Votes': 'sum',
    'Gross': 'sum',
    'Metascore': 'min'
})
```

```python
#passing list:
genres.agg(['mean', 'min', 'sum'])
```

```python
#adding both: dict + list
genres.agg(
{
    'Runtime':['mean', 'min'],
    'IMDB_Rating': 'mean',
    'No_of_Votes': ['sum','max'],
    'Gross': 'sum',
    'Metascore': 'min'
})
```

```python
#looping on groups.
df = pd.DataFrame(columns=movies.columns) # creating empty dataframe.
for group,data in genres:
    df = df.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])
    
df
```

```python
#split (apply) combine.
#apply -> builtin function
genres.apply(min)
```

```python
#find the number of movies starts with A for each group.
def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
#Find the rank of each movie in the group according to IMDB Score.
def rank_movies(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
genres.apply(rank_movies)
```

```python
#Find normalised IMDB group wise
def normal(group):
    group['norm_rating'] = group['IMDB_Rating'] - group['IMDB_Rating'].min()/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group

genres.apply(normal)
```

```python
ipl = pd.read_csv('deliveries.csv')
ipl.head()
```

```python
#Find the top 10 batsman in terms of runs.
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
#Find the batsman with max number of sixes.
six = ipl[ipl['batsman_runs']==6]
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
#Find the batsman with most number of 4's and 6's in last 5 overs.
temp_df = ipl[ipl['over']>15]
temp_df = temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6)]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
#Find the V kohli's record against all teams.
temp_df = ipl[ipl['batsman']=='V Kohli']
temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
#Create a function that returns the highest score of any batsman.
def highest(batsman):
    temp_df = ipl[ipl['batsman'] == batsman]
    return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('MS Dhoni')
```
