# What is Pandas ?

Pandas is a powerful, easy and flexible to use open source data analysis and manipulation tools built on the top of python.



# Pandas Series

A pandas series is like a column in table/dataset.It is a 1-D array holding data of any type.

# importing Pandas

```python
import numpy as np
import pandas as pd
```

# Series from list

```python
#string:
#Series object has index and its values.
country = ["India", "Pakistan", "USA", "Nepal", "Srilanka"]
pd.Series(country)

#0,1,2.. is the index numbers, 'India','USA',... are the values
#Pandas calls string dtype as 'object'.
```

```python
#Creating empty series
panda = []
type(pd.Series(panda))
```

```python
#integers:
runs = [13,24,56,78,100]
runs_series = pd.Series(runs)
runs_series
```

```python
#Custom Index:
#Default index can be replaced by our own custom index.
marks = [68,56,89,72,90]
subject = ['English', 'SST', "Science", "CS", "Maths"]

pd.Series(marks, index = subject) #making 'subject' as a index.
```

```python
#Setting a name:
#we can give a name to our series.
result = pd.Series(marks, index = subject, name = "Vedant's marksheet")
result
```

# Series from dictionary

```python
marks = {
    'English':68,
    'SST':56,
    'Science':89,
    'CS':72,
    'Maths':90
}

marks_series = pd.Series(marks, name = "Vedant's Marks")
marks_series
```

```python
temp = {
    0: 28,
    1: 43,
    2: 56
}
pd.Series(temp)
```
