```python
#Problem in vectorized operations in vanilla python.
s = ['cat','mat',None,'rat']
[i.startswith('c') for i in s]
#'None' is nothing here, and list comprehension gives error here.
```

```python
#How pandas solve this issue ?
s = pd.Series(['cat','mat',None,'rat'])
#String accessor.
s.str.startswith('c')
#Fast and optimized
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df['Name']
```

```python
#Common Functions:
#Lower/upper/capitalize/title
df['Name'].str.lower()
df['Name'].str.upper()
df['Name'].str.capitalize()
df['Name'].str.title()
```

```python
#len:
df['Name'].str.len()
```

```python
##len():
df['Name'].str.len().max()
```

```python
df['Name'][df['Name'].str.len()== 82].values[0]
```

```python
#strip:
"          Vedant          ".strip()
df['Name'].str.strip()
```

```python
#Split() -> get
df['lastname'] = df['Name'].str.split(',').str.get(0)
df.head()
```

```python
df[['title','firstname']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)
df.head()
```

```python
#replace:
df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
```

```python
df['title'].value_counts()
```

```python
#Filtering:
#Startswith()/endswith():
df[df['firstname'].str.startswith('A')]
```

```python
#isdigit/isalpha:
df[df['firstname'].str.isdigit()]
```

```python
#applying regex:
#contains:
#Search john,
df[df['firstname'].str.contains('john',case=False)].head()
```

```python
#find lastname with start and end char vowel.
df[df['lastname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')].head()
```

```python
df[df['lastname'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')].head()
```
