```python
#filling missing values (nan)
df['Investors Name'] = df['Investors Name'].fillna("Undisclosed")
sorted(df['Investors Name'].unique().tolist())
```

### Data Cleaning

```python
# Drop 'Remarks' Column.
df.drop(columns = ["Remarks"],inplace=True)
```

```python
df.head()
```

```python
# Setting Index
df.set_index('Sr No',inplace=True)
```

```python
df.head()
```

```python
#Renaming columns
df.rename(columns = {
    "Date dd/mm/yyyy": "date",
    "Startup Name": "startup",
    "Industry Vertical": "vertical",
    "SubVertical": "subvertical",
    "City  Location":"city",
    "Investors Name": "investors",
    "InvestmentnType": "round",
    "Amount in USD": "amount"
},inplace=True)
```

```python
df.head()
```

```python
# Changing USD to Rs in amount column

df['amount'] = df['amount'].fillna('0')
```

```python
df['amount'] = df['amount'].str.replace(',','')
df['amount'] = df['amount'].str.replace('undisclosed','0')
df['amount'] = df['amount'].str.replace('unknown','0')
df['amount'] = df['amount'].str.replace('Undisclosed','0')
```

```python
df = df[df['amount'].str.isdigit()]
```

```python
df['amount'] = df['amount'].astype('float')
```

```python
df.info()
```

```python
def to_inr(dollar):
    inr = dollar * 82.5
    return inr/10000000
```

```python
df['amount'] = df['amount'].apply(to_inr)
```

```python
df.head()
```

```python
# Converting 'date' column to datetime datatype
```

```python
df['date'] = df['date'].str.replace('05/072018','05/07/2018')
```

```python
df['date'] = pd.to_datetime(df['date'],errors='coerce')
```

```python
df.head()
```

```python
df.info()
```

```python
# Dropping missing values
df = df.dropna(subset=['startup','vertical','city','investors','round','amount'])
```

```python
df.head()
```

```python
df.to_csv("Startup_cleaned.csv",index=False)
```
